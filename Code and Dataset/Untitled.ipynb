```python
import pandas as pd
df = pd.read_csv('student.csv')
print(df.to_string())
```

```python
new_df = df.dropna()
print(new_df.to_string())
```

```python
for x in new_df.index:
  if ~(18 <= new_df.loc[x, 'Student_Age'] <= 25):
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
for x in new_df.index:
  if 'Male'!= new_df.loc[x, 'Sex'] != 'Female':
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
for x in new_df.index:
  if 'Bus'!= new_df.loc[x, 'Transportation'] != 'Private':
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
for x in new_df.index:
  if 'No'!= new_df.loc[x, 'Notes'] != 'Yes':
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
for x in new_df.index:
  if new_df.loc[x, 'Attendance'] != 'Sometimes' and new_df.loc[x, 'Attendance'] != 'Never' and  new_df.loc[x, 'Attendance'] != 'Always':
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
for x in new_df.index:
  if 'No'!= new_df.loc[x, 'Listening_in_Class'] != 'Yes':
    new_df.drop(x, inplace=True)
print(new_df.to_string())
```

```python
new_df.to_csv('CleanedFile.csv')
```
